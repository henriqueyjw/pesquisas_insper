### 3.2. Limpando a base de dados

Agora iremos somar somente as linhas que estão com dados nulos por coluna. Observe que usamos duas funções diferentes, a **<font color=blue>.isnull()</font>** indica somente se a observação contém ou não a informação em cada coluna, enquanto a função **<font color=blue>.sum()</font>** soma todas observações que não contém a informação em cada coluna. Caso altere a função **<font color=blue>.sum()</font>** por **<font color=blue>.mean()</font>** você terá a proporção de dados não preenchidos, ou seja, os números abaixo indicam quantos dados estão faltando em cada coluna.

```python
df.isnull().sum()
```

Caso queiramos retirar os dados faltantes apenas para uma ou algumas colunas, podemos utilizar a função **<font color=blue>.dropna()</font>** para identificar apenas as colunas que queremos que tenham todos os dados e retirar as observações que não contém essas informações.
No exemplo abaixo, retiramos todas as observações que não contém a informação da idade do passageiro, e somamos novamente as informações faltantes (note que agora a coluna **<font color=red>age</font>** não contém informação faltante).

```python
df.dropna(subset=['age']).isnull().sum()
```

Caso haja outras informações primordiais para a sua tarefa (informações que não podem faltar nos dados finais), você pode também indicar no mesmo campo que a coluna **<font color=red>age</font>** na função anterior.
```python
df.dropna(subset=['age', 'embarked']).isnull().sum()
```

### 3.3. Criando base de dados no `Python`

**"Até o momento, sabemos como abrir e tratar base de dados, mas como podemos fazer para criar uma base de dados e salvar ela para usar posteriormente ou em outro software?"**

Para isso, podemos usar também o pacote `Pandas`, mas com outra função. Primeiramente precisaremos relembrar os conceitos de lista e dicionário no `Python` (ambos foram abordados na apostila do Cactus que comentamos acima). 

No `Python` temos três tipos de variáveis muito úteis, as listas, tuplas e dicionários: 
- Uma **<font color=red>lista</font>** permite que os elementos sejam qualquer tipo de variável (até mesmo uma lista, tupla ou dicionário como elemento) e os elementos também podem ser diversos dentro da lista (diferentemente de softwares como o R), mas o principal ponto é que a lista permite que você inclua/exclua elementos em qualquer momento no seu script. 
- Uma **<font color=red>tupla</font>** se comporta igual uma lista quanto aos elementos, mas não permite que você altere os elementos ao longo do script, ou seja, ela é imutável. 
- Um **<font color=red>dicionário</font>** é basicamente uma lista mas que cada elemento é composto por uma <font color=red>chave</font> e um <font color=red>significado</font>, ou seja, cada elemento do dicionário é sempre composto por duas variáveis. Abaixo iremos mostrar as diferenças entre os três.

```python
lista = [1,2,3,4,'a','b','c','d']
tupla = (1,2,3,4,'a','b','c','d')
dicionario = {'a':1,'b':2,'c':3,'d':4}
```

```python
lista.append('final')
print('Nossa lista é:', lista)
```

```python
try:
    tupla.append('final')
except:
    print('Houve erro na inclusão de um elemento na tupla.')
```

```python
dicionario['e'] = 5
print('Nosso dicionário é:', dicionario)
```

**"Ok, mas o que essas coisas me ajudam a criar uma base de dados?"**

Lembra que uma base de dados é basicamente uma planilha? Então podemos definir cada coluna como se fosse uma <font color=red>chave</font> de um dicionário e o <font color=red>significado</font> ser o valor de cada uma das colunas para a observação que estamos olhando.

**"Ok, agora em português."** 

Para cada linha (observação) que temos na nossa base de dados, podemos colocar a característica dessa observação como o <font color=red>significado</font> de um dicionário aonde a <font color=red>chave</font> dela será a coluna com o nome da característica.
No exemplo abaixo, listamos características de várias pessoas, onde cada linha é uma pessoa e cada coluna é uma característica. Observe como iremos indicar para o `Python` como são os dados para cada pessoa.

```python
import pandas as pd
dados = list()
dicionario = {'nome':'Clara', 'altura':1.70,'peso':58,'curso':'direito'}
dados.append(dicionario)
dicionario = {'nome':'Mariana', 'altura':1.65,'peso':50,'curso':'administracao'}
dados.append(dicionario)
dicionario = {'nome':'Henrique', 'altura':1.73,'peso':65,'curso':'economia'}
dados.append(dicionario)
pd.DataFrame(dados)
```

Note que no exemplo anterior, cada característica é indicada como a <font color=red>chave</font> do dicionário e o valor dessa característica, para cada pessoa, é indicada como o <font color=red>significado</font> da chave. Outro ponto importante é que criamos uma lista de dicionários, onde cada elemento da lista é um dicionário com os dados de uma pessoa (observe a lista dados). Por fim, note que usamos uma função **<font color=blue>pd.DataFrame()</font>** para transformar essa lista de dicionários numa base de dados. Essa função interpreta cada linha como uma observação (nesse exemplo uma pessoa), cada <font color=red>chave</font> do dicionário como uma coluna e cada <font color=red>significado</font> como o valor correspondente àquela coluna para cada pessoa. Para executar qualquer operação com a base de dados resultante, seria necessário armazenar ela, atribuindo um 'nome'. No exemplo abaixo, iremos atribuir um nome para a base de dados e iremos salvar em dois formatos diferentes (**<font color=red>csv</font>** e em **<font color=red>excel</font>**).

```python
import os
path = '/mnt/e/natalia'
base = pd.DataFrame(dados)
base.to_csv(os.path.join(path, f'base_final.csv'),index=False) # Salva a base final em CSV

with pd.ExcelWriter(os.path.join(path, f'base_final.xlsx'),engine='openpyxl',mode='w') as writer:
    base.to_excel(writer,index = False) # Salva a base final em Excel
```

### 4.1. Navegando via código fonte

Nessa primeira forma, iremos navegar de uma forma mais simples (porém chata de ver). Em todos navegadores de internet, podemos entrar no código fonte da página da web que estamos. Lá ele identifica cada elemento que aparece na tela atribuindo nome e características para cada item.

Primeiro, vamos importar o pacote (aplicativo) que acessa a página. Alguns livros usam o pacote **<font color=blue>urllib</font>** e outros usam o **<font color=blue>requests</font>**. Usaremos o segundo por comodidade, mas no livro Python for Web Scrapping, os autores utilizam o primeiro pacote. Ambos fazem a mesma atividade. No exemplo abaixo, acessamos a página do imdb sobre um filme específico utilizando a função **<font color=blue>requests.get(url)</font>**, que busca o código fonte de uma página de internet(**<font color=red>url</font>**). Note que usamos a função **<font color=blue>.get(url)</font>**, pois nessa página não precisamos preencher nenhum dado para acessar as informações que queremos. 
Depois, conferimos o status da página com a função **<font color=blue>.status_code</font>**. Em geral, o status 200 indica que a página web foi encontrada.

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.imdb.com/title/tt5013056/'
response = requests.get(url)
response.status_code
```

Preferimos esse pacote pelo fato do `Python` armazenar o código fonte da página como um texto, sendo desnecessário abrir novamente a página todas vez. No exemplo abaixo, mostramos apenas os 5 mil primeiros caractéres do código fonte, para exemplificar como este aparece no `Python`, para isso utilizamos a função **<font color=blue>.text</font>** para obter o código fonte e selecionamos os caractéres pela forma apresentada anteriormente **<font color=blue>[:5000]</font>**.

```python
response.text[:500]
```

Para tornar o código fonte um pouco mais amigável de observar, podemos usar uma função do pacote  **<font color=blue>Beautiful Soup</font>** que serve para navegar dentro do código fonte das páginas e encontrar os objetos que você gostaria de capturar. Como o código fonte da página web pode ser enorme, iremos mostrar apenas os mil primeiro caracteres. Note que ainda assim continua sendo bem bagunçado e difícil de entender. O pacote que utilizamos transforma o código fonte da página web em um texto (string), onde o `Python`, utilizando outros pacotes, consegue identificar cada um dos elementos. No próximo exemplo, iremos salvar o código fonte desta página em uma variável e iremos utilizar o pacote (aplicativo) **<font color=blue>Beautiful Soup</font>** para identificar e navegar no código fonte. Para facilitar a visualização dos elementos, usamos o comando **<font color=blue>.prettify()</font>**, que indica a relação entre os elementos. Note que como o código fonte é enorme, mostramos apenas os mil primeiros catactéres.

```python
soup = BeautifulSoup(response.text, 'lxml')
soup.prettify()[:500]
```

**"Ok, mas como eu faço para saber aonde estão as informações que eu quero pegar?"**

Para saber qual <font color=red>tag</font> você deve buscar no código fonte, basta vc entrar na página de desenvolvedor do seu navegador. Lá, você vai olhar o código fonte e ele irá indicar na tela o que se refere cada linha do código. Supondo que eu quisesse a <font color=red>tag</font> do título do filme que estamos olhando no IMDB, eu poderia recorrer à <font color=red>tag</font> **<font color=red>title</font>** conforme o exemplo abaixo. Caso eu não indique qual característica dessa <font color=red>tag</font> eu quero ver, o `Python` irá me retornar todos os aspectos desse objeto. Caso eu quisesse apenas o texto do objeto, eu poderia indicar como **<font color=blue>.text</font>**, que ele retornaria apenas o conteúdo. No exemplo abaixo, procuramos todas as <font color=red>tags</font> **<font color=red>title</font>** com a função **<font color=blue>.find_all()</font>**.

```python
soup.find_all('title')
```

Nos próximos exemplos, iremos salvar algumas informações do filme que identificamos por meio de algumas <font color=red>tags</font>. Primeiro, iremos identificar todas as <font color=red>tags</font> existentes no código fonte, usando uma função que busca todas as <font color=red>tags</font> indicadas.
No exemplo abaixo, buscamos todas referentes à <font color=red>tag</font> **<font color=red>span</font>**, mas esse nome é muito comum, então repare que a lista que o `Python` retorna também contém várias informações. No exemplo abaixo, colocamos apenas os primeiros elementos da lista e a dimensão dela para mostrar a quantidade de elementos. No exemplo abaixo, procuramos todas as <font color=red>tags</font> **<font color=red>span</font>** com a função **<font color=blue>.find_all()</font>**, e mostramos na tela o total de elementos existentes no código fonte (**<font color=blue>len(target)</font>**) e alguns resultados dos <font color=red>tags</font> **<font color=red>span</font>** (**<font color=blue>target[:7]</font>**).

```python
target = soup.find_all("span")
print('A lista contém ',len(target), 'elementos.\n____________________________________________________________________________________________\n\n',target[:2])
```

Note que como a lista é muito extensa, podemos melhorar a busca do <font color=red>tag</font> que queremos, incluindo características dela na função. No exemplo abaixo, incluímos um nome atribuído à esse <font color=red>tag</font> que indica a nota do filme no IMDB, mas essa mesma função funciona com outras características da <font color=red>tag</font> como cor da fonte ou classe (**<font color=red>class</font>**). Com as funções que aprendemos até agora, podemos montar um script que busca também os links que você gostaria de procurar em uma página da web. No exemplo abaixo, entramos no site dos 250 filmes mais bem avaliados do IMDB e fizemos o próprio código buscar os links de cada filme para criar a nossa base de dados: 
- Entramos primeiro na página que enumera os filmes mais bem avaliados, buscamos todas as <font color=red>tag</font> **<font color=red>a</font>** (que geralmente indica algum link).
- Verificamos se a <font color=red>tag</font> tem algum link atrelado (**<font color=red>href</font>** nos atributos da <font color=red>tag</font> (**<font color=blue>'href' in item.attrs</font>**)), se o link tem algum nome atrelado (**<font color=red>title</font>** nos atributos da <font color=red>tag</font> (**<font color=blue>'title' in item.attrs</font>**)) e se o comprimento do link é igual à 17 (**<font color=blue>len(item.attrs['href'])==17</font>**), que indica que é um filme.

Note que estamos salvando cada link como um elemento de uma lista, esse passo é muito útil para montar a base de dados que queremos. Depois de obter todos links da página com os filmes mais bem avaliados, notamos que eles são apenas o complemento do link principal do IMDB, ou seja, temos que concatenar também o link da página principal com a função (**<font color=blue>'https://www.imdb.com'+item.attrs['href']</font>**), lembrando que ambos nesse caso são variáveis de texto (<font color=red>string</font>) e o `Python` concatena apenas com uma soma. Agora podemos repetir os passos que fizemos anteriormente para ter os dados de cada filme.

```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
lista = soup.find_all('a')

best_rated = list()
rows = list()

for item in lista:
    if 'href' in item.attrs and 'title' in item.attrs and len(item.attrs['href'])==17:
        best_rated.append('https://www.imdb.com'+item.attrs['href'])

for url in best_rated:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    dicionario = dict()
    dicionario['nome'] = soup.find_all('title')[0].text
    dicionario['nota'] = soup.find_all("span",{'itemprop':"ratingValue"})[0].text
    dicionario['publico'] = soup.find_all('span',{'itemprop':'ratingCount'})[0].text
    rows.append(dicionario)
    time.sleep(1)
pd.DataFrame(rows)
```

- Para percorrer a base de dados toda, iremos utilizar uma função chamada **<font color=blue>For</font>**, para identificar o conteúdo de cada linhae o índice de cada linha da base de dados iremos usar a função **<font color=blue>.iterrows()</font>** e por fim note que para incluir uma nova coluna, podemos criar uma lista com o total de itens igual o total de observações da base de dados (**<font color=red>saude</font>**) e incluir essa lista como uma nova coluna (**<font color=blue>df['saúde'] = saude</font>**).

```python
import os
import os.path
import re
import pandas as pd

ini_path = '/mnt/e/Mariana'
name_report = 'exemplo.csv'

df = pd.read_csv(os.path.join(ini_path, name_report))
df = df[1:]
saude = list()

for index,row in df.iterrows():
    if re.search(r'sa[uú]de|salud|health',row.texto,re.I):
        contagem = len(re.findall(r'sa[uú]de|salud|health',row.texto,re.I))
        saude.append(contagem)

df['saúde'] = saude
df.head()
```

**"Ok, e se eu quiser saber onde que aparece uma expressão no texto?"**

Para isso, podemos lembrar que como uma lista, tupla, dicionário ou base de dados, podemos selecionar um pedaço de um texto indicando o caractér inicial e o final, e o `Python` irá retornar todos caractéres entre os dois indicados. Esse quebra galho é muito útil para buscar trechos do texto. Supondo que você queira ver o trecho em que a palavra/expressão que você está procurando aparece no texto, podemos usar a função que busca a expressão (**<font color=blue>re.search()</font>**) e recuperar os caractéres aonde a expressão está. No exemplo abaixo, utilizamos a função **<font color=blue>re.search().span()</font>** para retornar uma lista com o primeiro e o último caractér da **primeira** expressão que se encaixa no padrão indicado. Depois disso verificamos se essa expressão está no início do texto ou no final do texto, com as condicionais **<font color=blue>Ifs</font>** e caso esteja no começo ou no final, consideramos o trecho desde o começo (**<font color=blue>inicial = 0</font>**) ou até o final do texto (**<font color=blue>final = len(row.texto)</font>**) e capturamos o trecho com a função **<font color=blue>row.texto[inicial:final]</font>**.

```python
import os
import os.path
import re
import pandas as pd

ini_path = '/mnt/e/Mariana'
name_report = 'exemplo.csv'

df = pd.read_csv(os.path.join(ini_path, name_report))
df = df[1:]
saude = list()

for index,row in df.iterrows():
    if re.search(r'sa[uú]de|salud|health',row.texto,re.I):
        caracteres = re.search(r'sa[uú]de|salud|health',row.texto,re.I).span()
        inicial = caracteres[0] - 50
        final = caracteres[1] + 50
        if inicial <0:
            inicial = 0
        if final > len(row.texto):
            final = len(row.texto)
        saude.append(row.texto[inicial:final])

df['trecho'] = saude
df.head()
```

**"Ok, e se eu quisesse verificar se dois ou mais termos aparecem no texto, sem importar a ordem entre eles?"**

Podemos montar um padrão que observa do início do texto até o final se há ou não os termos que estamos procurando. No exemplo abaixo buscaremos o termo relacionado à saúde e incluiremos um termo referente ao setor público, lembrando que essa lógica pode ser utilizada para mais de dois termos. Nos próximos exemplos iremos focar nos itens relacionados aos padrões indicados como indicadores de repetição, caractéres especiais, âncoras entre outros.

Nesse primeiro exemplo, a âncora indicada por **<font color=blue>^</font>** incida que o `Python` irá observar o texto do começo até o final. 

**"Ok, ele já não faz isso normalmente?"**

Sim, mas nesse caso, estamos perguntando se nos textos à frente (do início do texto) ele encontra algum termo relacionado à saúde (**<font color=blue>(?=.*sa[uú]de|.*salud|.*health)</font>**) ou à expressão público (**<font color=blue>(?=.*p[uú]blic)</font>**). Note que também colocamos outra condição para a busca (**<font color=blue>re.S</font>**), tornando o ponto final (**<font color=blue>.</font>**) no padrão indicado como qualquer caractér existente.

```python
import os
import os.path
import re
import pandas as pd

ini_path = '/mnt/e/Mariana'
name_report = 'exemplo.csv'

df = pd.read_csv(os.path.join(ini_path, name_report))
df = df[1:]
saude = list()

for index,row in df.iterrows():
    if re.search(r'^(?=.*sa[uú]de|.*salud|.*health)(?=.*p[uú]blic)',row.texto,re.I|re.S):
        saude.append('Sim')
    else:
        saude.append('Não')

df['Aparece os termos'] = saude
df.head()
```

**"Beleza, entendi que dentro dos padrões, podemos colocar várias coisas? Mas como eu faço quando tenho repetições ou variações de palavras?"**

Para ver todas os possíveis itens para usar nos padrões indicados, recomendo que você veja no site do pacote Regex para `Python` <https://docs.python.org/3/library/re.html> ou no site www.regex101.com, mas passarei pelos principais para auxiliar a montar os padrões mais simples.

Quanto temos algumas variações de palavras, podemos indicar as letras que podem mudar entre colchetes **<font color=red>\[ \]</font>**. O seguinte padrão **<font color=blue>pres[oa]</font>** captura os dois gêneros. 

**"Mas caso queiramos também incluir essas palavras no plural?"**

Podemos usar os caractéres que indicam repetição. Quando colocamos a quantidade de repetições entre chaves **<font color=red>\{ \}</font>**, caso seja um só numero ele obrigatoriamente terá que repetir aquele número de caracteres dentre as letras entre parênteses. Caso voê coloque dois números separados por uma vírgula, o `Python` entende que as repetições podem ser desde o primeiro número até o segundo. No seguinte exemplo temos **<font color=blue>pres[oas]{2}</font>** que obrigatoriamente vai capturar as palavras no plural enquanto o padrão  **<font color=blue>pres[oas]{1,2}</font>** captura tanto no singular como no plural.

**"Ok, e quando eu não souber o máximo de repetição ou quando temos pelo menos uma repetição?"**

Nesse caso, em vez de indicar com números entre chaves **<font color=red>\{ \}</font>**, podemos usar o ponto de interrogação (**<font color=red>?</font>**) para indicar zero ou uma repetição, o sinal de mais (**<font color=red>+</font>**) para indicar uma ou mais repetições ou o asterisco (**<font color=red>\*</font>**) para indicar de zero à infinitas repetições.

**"Beleza, e se eu não souber a letra, número ou tipo de espaço específico?"**

Também temos caractéres especiais para isso. Caso seja uma letra, usamos o **<font color=red>\w</font>** para indicar uma letra, o **<font color=red>\d</font>** para indicar um número e o **<font color=red>\s</font>** para indicar um espaço, enter ou algum outro tipo de espaçamento.

**"Hmm e se eu quiser capturar um trecho específico entre duas expressões?"**

Também conseguimos fazer isso, mas para isso temos que saber qual expressão aparece antes e qual aparece depois. Vamos chamar de Inicial e Final e podemos montar o seguinte padrão de captura **<font color=blue>(?<=Inicial)(.*?)(?=Final)</font>**, onde a primeira parte **<font color=blue>(?<=Inicial)</font>** pergunta se o que queremos capturar está depois da expressão inicial e a segunda parte pergunta se **<font color=blue>(?=Final)</font>** o que queremos está antes da expressão final. E o termo do meio **<font color=blue>(.*?)</font>** captura qualquer caractér, sendo palavra, número ou espaço em branco, mas tentando buscar menos informação possível (caso não coloquemos o ponto de interrogação (**<font color=red>?</font>**) depois do asterisco (**<font color=red>\*</font>**), o `Python` captura o máximo de informação possível, até mesmo o texto todo). O **<font color=blue>Regex</font>** sempre captura os termos que estão entre parênteses **<font color=red>\( \)</font>**, mas quando estamos lidando com uma palavra apenas, podemos usar somente o padrão sem eles. Nesse exemplo, iremos capturar todo que aparece entre as palavras relacionadas à saúde e à público **<font color=blue>(?<=health)(.*?)(?=p[uú]blic)</font>**.

```python
import os
import os.path
import re
import pandas as pd

ini_path = '/mnt/e/Mariana'
name_report = 'exemplo.csv'

df = pd.read_csv(os.path.join(ini_path, name_report))
df = df[1:]
saude = list()

for index,row in df.iterrows():
    if re.search(r'(?<=health)(.*?)(?=p[uú]blic)',row.texto,re.I|re.S):
        saude.append(re.search(r'(?<=health)(.*?)(?=p[uú]blic)',row.texto,re.I|re.S).groups()[0])
    else:
        saude.append('Não tem nada.')

df['Trecho entre Health e Public'] = saude
df.head()
```

#### 5.2.2 LDA e Topic Modeling

**"Beleza, mas teria como extrair algum padrão dos artigos que eu tenho?"**

Uma forma simples de extrair padrões é o *Topic Modeling* que computa estatísticamente as palavras que aparecem no mesmo com mais frequência e indica que elas podem estar relacionadas à um assunto (tópico). A forma mais comum de calcular é usando o Latent Dirichlet Allocation (LDA) que consiste numa distribuição binomial com N  entradas, sendo o N o total de palavras distintas que aparecem em todos os textos.

**"Ok, em português agora."**

O LDA assume que você pode ter várias palavras e cada uma delas tem uma probabilidade de aparecer ou não no texto, porém diferente de olhar para uma só palavra (distribuição binomial), olhamos para todas as palavras que achamos em todos os textos que estamos estudando (distribuição Dirichlet).

**"Beleza, mas o que esse método assume?"**

O LDA se encaixa numa família de modelos chamada "bag of words", que não considera o contexto em que as palavras se encontram (semântica) e observa apenas a frequência delas nos textos que estamos estudando.

**"Puts, mas isso não atrapalha?"**

Numa primeira análise não. Pois você consegue extrair padrões de acordo com as frequências de palavras. Mas caso você queira entender mais a fundo as informações dos textos, talvez seja mais apropriado outros modelos que consideram também a semântica do texto (próximo tópico).

Como o `Python` possui diversos desenvolvedores e aplicativos diferentes, em alguns casos algumas coisas acabam tendo mais de um método para fazer. No caso do LDA temos dois métodos com aplicativos muito usuais em Machine Learning. Iremos apresentar os dois, partindo do mais simples para o mais complexo. Antes disso iremos mostrar como definir uma função no `Python` e utilizar ela ao longo do script. Esses métodos ajudam bastante quando queremos simplificar o script com entradas e saídas bem definidas de cada função.

Para definir uma função, precisamos indicar um nome para ela e quais variáveis que ela usará. No exemplo abaixo definimos uma função chamada **<font color=red>teste</font>** e que utiliza as variáveis **<font color=red>var_um</font>** e **<font color=red>var_dois</font>**. Essa função irá multiplicar o <font color=red>var_um</font> pelo <font color=red>var_dois</font> (uma coisa engraçada é que se o <font color=red>var_um</font> for um texto a função retorna o texto várias vezes, e se for um número, ela retorna o resultado da multiplicação). Para "chamar" a função, iremos usar o comando **<font color=blue>teste(var_um,var_dois)</font>**, o comando **<font color=blue>return</font>** indica o que a função irá retornar quando ela for chamada, no exemplo ela pode retornar tanto um texto como um número.

Definimos a função que calcula e apresenta as palavras mais frequentes com os seguintes passos:
- A função irá receber uma matriz <font color=red>count_data</font>, onde cada linha é um dos textos que estamos estudando e cada coluna é a frequência de uma palavra específica. Esse método mapeia todas as palavras que aparecem em todos os textos. 
- A função também recebe o método que usamos para vetorizar os textos <font color=red>count_vectorizer</font>;
- Na análise que usamos essa função, estavamos utilizando para vários períodos diferentes, então indicamos o ano em que terminava a análise também para a função;

A função inicia retirando as palavras mapeadas pelo método de vetorização <font color=red>word</font> e cria um vetor de zeros que irá mapear a frequência de cada palavra <font color=red>total_counts</font>;
É definido um loop <font color=red>for</font> para computar a frequência de cada palavra em cada texto, somando a frequência dessa palavra em cada texto;
É criado uma matriz que indica cada palavra e a sua frequência nos textos que estamos estudando <font color=red>count_dict</font> e depois reclassificamos a matriz de acordo com a frequência das palavras (maior para o menor);
Por fim, definimos as características do gráfico e criamos o gráfico na tela do `Python`, poderíamos também salvar o gráfico em imagem (o código que está como comentário salva o gráfico).

```python
def plot_10_most_common_words(count_data, count_vectorizer,ano):
    import matplotlib.pyplot as plt
    words = count_vectorizer.get_feature_names()
    total_counts = np.zeros(len(words))
    for t in count_data:
        total_counts+=t.toarray()[0]
    
    count_dict = (zip(words, total_counts))
    count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:10]
    words = [w[0] for w in count_dict]
    counts = [w[1] for w in count_dict]
    x_pos = np.arange(len(words)) 
    
    plt.figure(2, figsize=(15, 15/1.6180))
    plt.subplot(title= '10 most common words')
    sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.5})
    sns.barplot(x=x_pos, y=counts, palette='husl')
    plt.xticks(x_pos, words, rotation=90) 
    plt.xlabel('words')
    plt.ylabel('counts')
    plt.show()
#     plt.savefig(os.path.join(root,f'freq_{ano}.png'))
```

Agora iremos definir a função que apresenta as palavras relacionadas a cada tópico e os principais tópicos entre os textos que estamos estudando. Essa função irá receber: 
- O modelo LDA do pacote SK-Learn já calculado para a nossa base de dados <font color=red>model</font>;
- O modelo que estamos utilizando para vetorizar os dados <font color=red>count_vectorizer</font>; 
- O total de palavras que queremos mostrar <font color=red>n_top_words</font>.

A função inicia retirando as palavras mapeadas pelo modelo de vetorização utilizado e utiliza um loop <font color=red>for</font> para observar o tópico e as palavras relacionadas à cada tópico que o modelo mapeou. Dentro desse loop, a função apresenta o número do tópico e as palavras relacionadas à esse tópico.

```python
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

Por fim, podemos definir o total de tópicos que queremos mapear e a quantidade de palavras relacionadas à cada tópico, criar o modelo e imprimir as palavras relacionadas à cada tópico.

```python
number_topics = 5 # Total de tópicos a serem mapeados pelo modelo
number_words = 10 # Total de palavras para cada tópico

lda = LDA(n_components=number_topics, n_jobs=-1) # Definindo o modelo LDA a ser mapeado
lda = lda.fit(count_data) # Mapeando os dados para os dados que já vetorizados

print_topics(lda, count_vectorizer, number_words) # Chamando a função que imprime os tópicos
```

Definimos uma função que recebe uma <font color=red>lista de textos</font>, transforma cada um dos textos em uma <font color=red>lista</font> de palavras (no caso desconsiderando a pontuação) e retorna uma  <font color=red>lista</font> com todos textos em forma de  <font color=red>lista</font> (uma **<font color=red>lista de listas</font>**).

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations
```

Definimos uma função que recebe uma **<font color=red>lista de listas</font>** com todos os textos que estamos analisando, transforma cada texto (em <font color=red>lista</font>) de volta para um <font color=red>texto</font>, aplica o pre-processamento do pacote Gensim e avalia se alguma palavra desse texto se encaixa nos conectores indicados (<font color=red>stop_words</font>) e por fim, retorna uma **<font color=red>lista de listas</font>** com todos os textos sem os conectores.

```python
# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]
```

Definimos uma função que mapeia todos os bigramas (conjunto de duas palavras) que aparecem frequêntemente juntos (pelo menos 5 vezes).

```python
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]
```

Definimos uma função que mapeia todos os trigramas (conjunto de três palavras) que aparecem frequêntemente juntos.

```python
def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]
```

Por fim, definimos uma função que recebe uma **<font color=red>lista de listas</font>** e as palavras que serão aceitas nas análise que queremos.

```python
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

Retiramos os conectores, juntamos as palavras dos bigramas e trigramas, definimos o leitor que classifica as palavras sintaticamente (em verbos, advérbio, sujeito e afins) e retiramos apenas os tipos de palavras que temos interesse em estudar.

Depois disso, definimos o nosso modelo LDA, apresentamos os principais tópicos, a relevância de cada palavra no tópico, a perplexidade (médida de qualidade do LDA) e a coerência dos tópicos mapeados.

A medida de entropia verifica o quanto o texto apresenta diversidade de palavras, quanto maior a medida maior será a diversidade de palavras que o texto apresenta. Caso a medida seja próxima de zero, temos poucas palavras com muita frequência. No exemplo abaixo calculamos a entropia para os 10 primeiro artigos da nossa base de dados.

```python
import math

def entropy(freq):
    total = sum(freq.values())
    H = 0.0
    for c in freq.values():
        prob = c/total
        H -= prob*math.log2(prob)
    return H

data = df [:10]
for index,row in data.iterrows():
    text = row.texto.lower().split()
    freq = nltk.FreqDist(text)
    print(f'{index}: {entropy(freq):3f} bits')
```

**"Mas esse cálculo serve para algo mais fora isso?"**

Claro! A idéia obtida com a entropia é uma parte central dos modelos Naïve-Bayes. Esses modelos de linguagem se baseiam nas probabilidades de aparição de cada palavra, ou seja, nas frequências de cada palavra.
No exemplo abaixo iremos definir o tamanho do vocabulário dos textos que estamos analisando e depois iremos calcular a probabilidade de aparecer uma frase que queremos observar.

Note que para montar esse cálculo de probabilidade, definimos uma classe. Essa forma de programação é um pouco diferente do que estamos acostumados até aqui. Irei explicar com exmplos práticos essa forma nos capítulos seguintes. Nesse momento, apenas observe que usamos a diversidade de palavras de todos os textos que estamos estudando (contagem de todas as diferentes palavras que aparecem nos textos) para padronizar o cálculo da nossa probabilidade, pois os valores tendem a ficar muito próximos à zero.

Nesse exemplo, iremos inserir um texto (em formato de lista) na nossa função e indicar alguma sentença que queremos ver a probabilidade dela aparecer no texto que inserimos. No caso, coloquei uma frase famosa de um filme e duas palavras frequêntes nos textos ligados à saúde.

```python
import math

alpha = 0.1
corpus = list()
for index,row in df.iterrows():
    corpus = corpus + row.texto.lower().split()
Vocab = len(set(corpus))

class ProbSentenceNaive:
    def __init__(self, freq, alpha, V):
        N = sum(freq[w] for w in freq)
        log_N_smooth = math.log10(N + alpha*V)
        log_prob = {}
        for w in freq:
            log_prob[w] = math.log10(freq[w] + alpha) - log_N_smooth
        self.log_prob = log_prob
        self.default_prob = math.log10(alpha) - log_N_smooth
        
    def __call__(self, sentence):
        freq_sentence = nltk.FreqDist(sentence.split())
        log_prob_sentence = 0.0
        for w in freq_sentence:
            this_log_prob = self.log_prob[w] if w in self.log_prob else self.default_prob
            log_prob_sentence += freq_sentence[w] * this_log_prob
        return log_prob_sentence

freq = nltk.FreqDist(df.texto[0].lower().split())
prob_sentenca_naive = ProbSentenceNaive(freq, alpha, Vocab)
print("O log likelihood da sentença 'may the force be with you' é", prob_sentenca_naive('may the force be with you'))
print("O log likelihood da sentença 'right to health' é", prob_sentenca_naive('right to health'))
```

No exemplo acima, quanto maior o log likelihood, maior é a probabilidade da sentença aparecer no texto que indicamos.

**"Beleza, mas essa probabilidade me ajuda com os dados que tenho?"**

Supondo que você, ou algum conhecido, já tenham estudado uma base de dados e classificado ela de acordo com algum critério como (de interesse/não interesse; positivo/negativo; direito/saúde/ambos). Agora, podemos montar uma forma de classificar novos textos, utilizando os textos já classificados como dados de entrada para o nosso modelo. Esse modelo de classificação é chamado de Naïve-Bayes, porque observa apenas as probabilidades dos textos/sentenças serem próximos dos já temos classificados.

No exemplo abaixo, calculamos novamente o likelihood (probabilidade) da sentença aparecer para cada um dos 10 primeiros textos da nossa base de dados, e depois apresentamos na ordem do texto mais provável ao mais improvável que a sentença aparece.

**"Hein? Não entendi."**

Nesse exemplo, primeiro fizemos os passos anteriores para achar o vocabulário total dos textos que estamos estudando <font color=red>Vocab</font>; Depois calculamos para cada um dos 10 primeiros textos, a probabilidade da sentença que indicamos aparecer nesse texto; Por fim, enumerados o texto com maior probabilidade para o texto com menor probabilidade.

```python
import math

alpha = 0.1
corpus = list()
for index,row in df.iterrows():
    corpus = corpus + row.texto.lower().split()
Vocab = len(set(corpus))

lm = {}
data = df [:10]
for index,row in data.iterrows():
    text = row.texto.lower().split()
    freq = nltk.FreqDist(text)
    lm[index] = ProbSentenceNaive(freq, alpha, Vocab)
    
sentenca = 'right to health'

likelihoods = []
for fileid, model in lm.items():
    likelihoods.append((model(sentenca), fileid))

for like, title in sorted(likelihoods, reverse=True):
    print(f'{like:.2f}\t{title}')
```

Utilizando a função de vetorização do pacote **<font color=blue>SK Learn</font>**, separamos da nossa base de dados, o conjunto de textos classificados como **<font color=red>statement</font>** e vetorizamos as frases. Abaixo mostramos como fica cada um dos textos depois de vetorizados.

```python
q_docs = [row['sent'] for index,row in training_data.iterrows() if row['class'] == 'question']

vec_q = CountVectorizer()
X_q = vec_q.fit_transform(q_docs)
tdm_q = pd.DataFrame(X_q.toarray(), columns=vec_q.get_feature_names())
tdm_q
```

Agora podemos ver a frequência de cada uma das palavras nos dois conjuntos de dados.

```python
word_list_s = vec_s.get_feature_names();    
count_list_s = X_s.toarray().sum(axis=0) 
freq_s = dict(zip(word_list_s,count_list_s))
print('As frequências para o conjunto statement são',freq_s)

word_list_q = vec_q.get_feature_names();    
count_list_q = X_q.toarray().sum(axis=0) 
freq_q = dict(zip(word_list_q,count_list_q))
print('As frequências para o conjunto question são',freq_q)
```

Agora podemos contar quantas palavras e a frequencia de cada na base de dados toda e o total de palavras em cada um dos conjuntos que temos **<font color=red>question</font>** e **<font color=red>statement</font>**.

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = [row['sent'] for index,row in training_data.iterrows()]
vec = CountVectorizer()
X = vec.fit_transform(docs)

total_features = len(vec.get_feature_names())
total_cnts_features_s = count_list_s.sum(axis=0)
total_cnts_features_q = count_list_q.sum(axis=0)
```

Indicamos a frase que queremos classificar e transformamos a frase numa lista de palavras.

```python
from nltk.tokenize import word_tokenize
new_sentence = 'what is the price of the book'
new_word_list = word_tokenize(new_sentence)
```

Por fim, calculamos a probabilidade dessa frase ser uma **<font color=red>statement</font>**.

```python
prob_s_with_ls = []
final_prob = 0.5
for word in new_word_list:
    if word in freq_s.keys():
        count = freq_s[word]
    else:
        count = 0
    prob_s_with_ls.append((count + 1)/(total_cnts_features_s + total_features))
    final_prob = final_prob*(count + 1)/(total_cnts_features_q + total_features)
dict(zip(new_word_list,prob_s_with_ls))
print('A probabilidade da frase ser uma statement é',final_prob)
```

E calculamos a probabilidade dessa frase ser uma **<font color=red>question</font>**.

```python
prob_q_with_ls = []
final_prob = 0.5
for word in new_word_list:
    if word in freq_q.keys():
        count = freq_q[word]
    else:
        count = 0
    prob_q_with_ls.append((count + 1)/(total_cnts_features_q + total_features))
    final_prob = final_prob*(count + 1)/(total_cnts_features_q + total_features)
dict(zip(new_word_list,prob_q_with_ls))
print('A probabilidade da frase ser uma question é',final_prob)
```

Note que a probabilidade dela ser uma **<font color=red>question</font>** é maior que ser uma **<font color=red>statement</font>**, logo, classificamos ela como uma **<font color=red>question</font>**.

**"Ok, esse modelo considera apenas as probabilidade das palavras, correto? E se alguém falasse igual o mestre Yoda?"**


Obs: Para quem não conhece, mestre Yoda é um dos personagens icônicos da saga Star Wars, que troca a ordem das palavras em suas falas, usando primeiro o predicado para depois falar o sujeito.

Nesse caso, o modelo que temos atualmente não consegue distinguir alguem que fala igual o mestre Yoda de alguém que fala normalmente, pois o Naïve-Bayes não considera a ordem em que as palavras aparecem. Caso tenhamos que considerar isso, podemos construir um modelo de bigramas ou trigramas (conjuntos de duas ou três palavras), onde a ordem das palavras importa.

Nesse primeiro exemplo, primeiro quebramos a sentença que estamos estudando em diversos conjuntos de duas palavras e calculamos a probabilidade delas nos textos que temos classificados.

```python
from nltk.util import bigrams

corpus = list()
for index,row in df.iterrows():
    corpus = corpus + row.texto.lower().split()
Vocab_bigram = len(set(bigrams(corpus)))

class ProbSentenceNaive:
    def __init__(self, freq, alpha, V):
        N = sum(freq[w] for w in freq)
        log_N_smooth = math.log10(N + alpha*V)
        log_prob = {}
        for w in freq:
            log_prob[w] = math.log10(freq[w] + alpha) - log_N_smooth
        self.log_prob = log_prob
        self.default_prob = math.log10(alpha) - log_N_smooth
        
    def __call__(self, sentence):
        freq_sentence = nltk.FreqDist(list(bigrams(sentence)))
        log_prob_sentence = 0.0
        for w in freq_sentence:
            this_log_prob = self.log_prob[w] if w in self.log_prob else self.default_prob
            log_prob_sentence += freq_sentence[w] * this_log_prob
        return log_prob_sentence

lm = {}
data = df [:10]
for index,row in data.iterrows():
    text = list(bigrams(row.texto.lower().split()))
    freq = nltk.FreqDist(text)
    lm[index] = ProbSentenceNaive(freq, alpha, Vocab_bigram)
    
sentenca = 'right to health'

likelihoods = []
for fileid, model in lm.items():
    likelihoods.append((model(sentenca), fileid))

for like, title in sorted(likelihoods, reverse=True):
    print(f'{like:.2f}\t{title}')
```

Nesse próximo exemplo, quebramos a sentença em conjuntos de três palavras e calculamos as probabilidades desses conjuntos nos textos que já temos classificados.

```python
from nltk.util import trigrams

corpus = list()
for index,row in df.iterrows():
    corpus = corpus + row.texto.lower().split()
Vocab_trigram = len(set(trigrams(corpus)))

class ProbSentenceNaive:
    def __init__(self, freq, alpha, V):
        N = sum(freq[w] for w in freq)
        log_N_smooth = math.log10(N + alpha*V)
        log_prob = {}
        for w in freq:
            log_prob[w] = math.log10(freq[w] + alpha) - log_N_smooth
        self.log_prob = log_prob
        self.default_prob = math.log10(alpha) - log_N_smooth
        
    def __call__(self, sentence):
        freq_sentence = nltk.FreqDist(list(trigrams(sentence)))
        log_prob_sentence = 0.0
        for w in freq_sentence:
            this_log_prob = self.log_prob[w] if w in self.log_prob else self.default_prob
            log_prob_sentence += freq_sentence[w] * this_log_prob
        return log_prob_sentence

lm = {}
data = df [:10]
for index,row in data.iterrows():
    text = list(trigrams(row.texto.lower().split()))
    freq = nltk.FreqDist(text)
    lm[index] = ProbSentenceNaive(freq, alpha, Vocab_bigram)
    
sentenca = 'right to health'

likelihoods = []
for fileid, model in lm.items():
    likelihoods.append((model(sentenca), fileid))

for like, title in sorted(likelihoods, reverse=True):
    print(f'{like:.2f}\t{title}')
```

A próxima forma de representar os dados é o método chamado tf-idf, que também haviamos comentado anteriormente. Esse método computa a frequência da palavra nos textos e padroniza pela frequência de textos que apresentam essa palavra em relação à todos os textos estudados.

No exemplo abaixo, complementamos a base de dados com a contagem de elementos do texto como nomes, pronomes, verbos, adjetivos e advérbios:

```python
pos_family = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}

# function to check and get the part of speech tag count of a words in a given sentence
def check_pos_tag(x, flag):
    cnt = 0
    try:
        wiki = textblob.TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_family[flag]:
                cnt += 1
    except:
        pass
    return cnt

trainDF['noun_count'] = trainDF['text'].apply(lambda x: check_pos_tag(x, 'noun'))
trainDF['verb_count'] = trainDF['text'].apply(lambda x: check_pos_tag(x, 'verb'))
trainDF['adj_count'] = trainDF['text'].apply(lambda x: check_pos_tag(x, 'adj'))
trainDF['adv_count'] = trainDF['text'].apply(lambda x: check_pos_tag(x, 'adv'))
trainDF['pron_count'] = trainDF['text'].apply(lambda x: check_pos_tag(x, 'pron'))
trainDF.head()
```

#### Criação de modelos de classificação

Agora que já vimos várias formas diferentes de inserir os textos de treinamento, podemos começar a criar os nossos modelos de classificação. Iremos apresentar diversos modelos e o grau de acerto deles para cada tipo de representação dos textos inseridos (textos de treinamento). Primeiro vamos criar uma função que retorna a acurácia do modelo de classificação:

```python
def train_model(classifier, feature_vector_train, label, feature_vector_valid, is_neural_net=False):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)
    
    if is_neural_net:
        predictions = predictions.argmax(axis=-1)
    
    return metrics.accuracy_score(predictions, valid_y)
```

#### K-Nearest Neighborhood

**"Ok, entendi que existem vários modelos para classificar os textos de acordo com as características ou palavras dos textos. Há mais algum que eu deveria saber?"**

O último modelo desse gênero que iremos mostrar é o  K nearest neighborhood ou KNN, que busca os textos com vetores mais próximos para classificar o que estamos interessado. Nesse site, os autores fazem um exemplo simples de KNN que iremos analisar passo a passo <https://towardsdatascience.com/text-classification-using-k-nearest-neighbors-46fa8a77acc5>. Esse modelo verifica quais textos possuem características similares em uma determinada "distância" entre esses vetores de características. Lembra nas aulas de matemática, quando tinhamos que calcular a distância entre dois pontos em planos 2D (x e y)? Isso é muito parecido, iremos definir uma distância na qual iremos comparar os vetores próximos ao que estamos classificando e contar quantos possuem o rótulo 0 e o rótulo 1. 

No exemplo abaixo, primeiro vamos importar os pacotes (aplicativos) que iremos usar na nossa análise.

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.corpus import genesis
# nltk.download('genesis')
# nltk.download('wordnet')
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
genesis_ic = wn.ic(genesis, False, 0.0)

import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.corpus import stopwords
from sklearn.metrics import roc_auc_score
```

Primeiro vamos criar o classificador, definindo uma classe. Nela, indicamos o K e o tipo de distância. 
- Quando usamos a função **<font color=red>.fit()</font>**, indicamos os dados de teste e os rótulos desses dados. 
- Para usar a função **<font color=red>.predict()</font>**, indicamos os dados de teste e o modelo irá retornar as previsões do modelo de acordo com os K vetores próximos ao que estamos classificando.

Criamos também algumas funções para tratar os textos antes de inserir eles no modelo. 
- A função **<font color=blue>convert_tag(self, tag)</font>** serve para converter os tags de uma função de classificação de palavras no pacote **<font color=blue>nltk</font>** para outro pacote (o **<font color=blue>synsets</font>**). 
- A função **<font color=blue>doc_to_synsets(self, doc)</font>** converte os textos para a representação usada no pacote **<font color=blue>synsets</font>**.

Por fim criamos as funções que calculam a similaridade das frases e dos documentos.
- A função **<font color=blue>similarity_score(self, s1, s2, distance_type = 'path')</font>** calcula a similaridade de dois textos quaisquer.
- A função **<font color=blue>document_similarity(self,doc1, doc2)</font>** calcula a similaridade de dois documentos

```python
class KNN_NLC_Classifer():
    def __init__(self, k=1, distance_type = 'path'):
        self.k = k
        self.distance_type = distance_type

    # This function is used for training
    def fit(self, x_train, y_train):
        self.x_train = x_train
        self.y_train = y_train

    # This function runs the K(1) nearest neighbour algorithm and
    # returns the label with closest match. 
    def predict(self, x_test):
        self.x_test = x_test
        y_predict = []

        for i in range(len(x_test)):
            max_sim = 0
            max_index = 0
            for j in range(self.x_train.shape[0]):
                temp = self.document_similarity(x_test[i], self.x_train[j])
                if temp > max_sim:
                    max_sim = temp
                    max_index = j
            y_predict.append(self.y_train[max_index])
        return y_predict
    
    def convert_tag(self, tag):
    #     Convert the tag given by nltk.pos_tag to the tag used by wordnet.synsets
        tag_dict = {'N': 'n', 'J': 'a', 'R': 'r', 'V': 'v'}
        try:
            return tag_dict[tag[0]]
        except KeyError:
            return None

    def doc_to_synsets(self, doc):

    #         Returns a list of synsets in document.
    #         Tokenizes and tags the words in the document doc.
    #         Then finds the first synset for each word/tag combination.
    #     If a synset is not found for that combination it is skipped.

    #     Args:
    #         doc: string to be converted

    #     Returns:
    #         list of synsets

        tokens = word_tokenize(doc+' ')

        l = []
        tags = nltk.pos_tag([tokens[0] + ' ']) if len(tokens) == 1 else nltk.pos_tag(tokens)

        for token, tag in zip(tokens, tags):
            syntag = self.convert_tag(tag[1])
            syns = wn.synsets(token, syntag)
            if (len(syns) > 0):
                l.append(syns[0])
        return l 
    
    def similarity_score(self, s1, s2, distance_type = 'path'):
    #       Calculate the normalized similarity score of s1 onto s2
    #       For each synset in s1, finds the synset in s2 with the largest similarity value.
    #       Sum of all of the largest similarity values and normalize this value by dividing it by the
    #       number of largest similarity values found.

    #       Args:
    #           s1, s2: list of synsets from doc_to_synsets

    #       Returns:
    #           normalized similarity score of s1 onto s2

        s1_largest_scores = []

        for i, s1_synset in enumerate(s1, 0):
            max_score = 0
            for s2_synset in s2:
                if distance_type == 'path':
                    score = s1_synset.path_similarity(s2_synset, simulate_root = False)
                else:
                    score = s1_synset.wup_similarity(s2_synset)                  
                if score != None:
                    if score > max_score:
                        max_score = score

            if max_score != 0:
                s1_largest_scores.append(max_score)

        mean_score = np.mean(s1_largest_scores)

        return mean_score  

    def document_similarity(self,doc1, doc2):
    #           Finds the symmetrical similarity between doc1 and doc2
        synsets1 = self.doc_to_synsets(doc1)
        synsets2 = self.doc_to_synsets(doc2)

        return (self.similarity_score(synsets1, synsets2) + self.similarity_score(synsets2, synsets1)) / 2
```

**"Hmm, tem algum exemplo mais simples para explicar essa coisa de função dentro de classe?"**

Temos sim, vamos montar um aqui abaixo que pode ajudar nisso. Definimos uma classe genérica que recebe uma informação quando indicamos ao criar o objeto dessa classe, e automaticamente o `Python` busca que tipo de informação estamos colocando (um número inteiro/decimal ou um texto). Depois criamos as funções **<font color=blue>sum()</font>**, que só soma números e reclama quando temos texto; **<font color=blue>mult()</font>** que só multiplica números e reclama quando temos texto e a função **<font color=blue>caracter()</font>** que conta a quantidade de caractéres de um texto e reclama quando temos um número.

```python
class objeto():
    def __init__(self, information):
        self.informacao = information
        self.tipo = type(information)
    
    def sum(self,var):
        if self.tipo == str:
            print('Pô, essa informação não pode ser somada!')
        elif self.tipo == int or self.tipo == float:
            print(self.informacao + var)
    
    def mult(self,var):
        if self.tipo == str:
            print('Pô, essa informação não pode ser multiplicada!')
        elif self.tipo == int or self.tipo == float:
            print(self.informacao * var)
            
    def caracter(self):
        if self.tipo == str:
            print(len(self.informacao))
        elif self.tipo == int or self.tipo == float:
            print('Pô, não contamos caractéres de números!')
```

Abaixo testamos a nossa classe e as funções inserindo um texto e depois um número.

```python
dado = objeto('Shit')
dado.sum(5)
dado.mult(5)
dado.caracter()
```

```python
dado = objeto(25)
dado.sum(5)
dado.mult(5)
dado.caracter()
```
